## Block size testing

```python
## optimal block size of global signal (average across all parcelations)
bs_globle=np.load('blocksize_sub_globlesig.npy')
bs_globle.shape
fig, axs = plt.subplots(1,4,figsize=(16,4),dpi=600)
for i in range(0,4):
    sns.distplot(bs_globle[i,:],kde=False,ax=axs[i])
```

```python
## optimal block size per parcel per subjects

data=np.load('blocksize_sub_parcel_10.npy')
data.shape

fig, axs = plt.subplots(2,4,figsize=(16,8),dpi=600)
for i in range(4,12):
    subdata=np.reshape(data[:,i,:],(4,998))
    sub_corr=np.corrcoef(subdata)
    plotidx=axs[int(np.floor((i-4)/4)),int((i-4)%4)]
    sns.heatmap(sub_corr,square=True,annot=True, fmt='.2g',xticklabels=['5','10','30','60'],yticklabels=['5','10','30','60'],ax=plotidx)
```
